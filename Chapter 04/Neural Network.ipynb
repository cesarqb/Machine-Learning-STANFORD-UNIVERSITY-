### Model Representation I

Examinemos cómo representaremos una función de hipótesis utilizando redes neuronales. En un nivel muy simple, las neuronas son básicamente unidades computacionales que toman entradas (dendritas) como entradas eléctricas (llamadas "picos") que se canalizan a salidas (axones). En nuestro modelo, nuestras dendritas son como las características de entrada, y la salida es el resultado de nuestra función de hipótesis. En este modelo, nuestro nodo de entrada a veces se denomina "unidad de polarización". Siempre es igual a 1. En las redes neuronales, utilizamos la misma función logística que en la clasificación, frac {1} {1 + e ^ {- \ theta ^ Tx}} 1 + e −θ T x 1 , sin embargo, a veces lo llamamos una función de activación sigmoidea (logística). En esta situación, nuestros parámetros "theta" a veces se denominan "pesos".



Nuestros nodos de entrada (capa 1), también conocidos como la "capa de entrada", van a otro nodo (capa 2), que finalmente genera la función de hipótesis, conocida como la "capa de salida".

Podemos tener capas intermedias de nodos entre las capas de entrada y salida llamadas "capas ocultas".

En este ejemplo, etiquetamos estos nodos de capa intermedios u "ocultos" y los llamamos "unidades de activación".


Esto significa que calculamos nuestros nodos de activación mediante el uso de una matriz de parámetros 3 × 4. Aplicamos cada fila de los parámetros a nuestras entradas para obtener el valor para un nodo de activación. El resultado de nuestra hipótesis es la función logística aplicada a la suma de los valores de nuestros nodos de activación, que se han multiplicado por otra matriz de parámetros \ Theta ^ {(2)} Θ que contiene los pesos para nuestra segunda capa de nodos.
